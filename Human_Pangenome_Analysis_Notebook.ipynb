# ðŸ§¬ Human Pangenome Reference Analysis

This notebook explores key insights from the research paper **"A Draft Human Pangenome Reference" (Nature, 2023)** using simulated datasets.

It demonstrates how diverse genomic references improve variant detection and mapping accuracy, especially for underrepresented populations. The notebook includes:

- ðŸ“Š Visualization of **structural variant counts** across global populations
- ðŸ“ˆ Comparison of **read mapping accuracy**: GRCh38 vs Pangenome
- ðŸ§  A **PCA plot** illustrating genetic variation between populations

The goal is to showcase the importance of inclusive genome references in precision medicine and bioinformatics using clean, reproducible analysis.

**Keywords:** Human Pangenome, Bioinformatics, Structural Variants, PCA, Population Genomics, Precision Medicine


```python
# Data manipulation and analysis
import pandas as pd
import numpy as np

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Machine Learning (for PCA)
from sklearn.decomposition import PCA

# Set a clean plot style
sns.set(style="whitegrid")
```

```python
# Load structural variant data
df_sv = pd.read_csv('structural_variant_counts.csv')

# Load mapping accuracy data
df_map = pd.read_csv('mapping_accuracy.csv')

# Load PCA data
df_pca = pd.read_csv('pca_population_variation.csv')
```

```python
# Preview data
print("SV Data:")
print(df_sv.head())

print("\nMapping Accuracy Data:")
print(df_map.head())

print("\nPCA Data:")
print(df_pca.head())

# Check for missing values
print("\nMissing values check:")
print("SV:", df_sv.isnull().sum().sum())
print("Mapping:", df_map.isnull().sum().sum())
print("PCA:", df_pca.isnull().sum().sum())

# Convert types if necessary
df_sv['SV_Count'] = df_sv['SV_Count'].astype(int)
df_map['Mapping_Accuracy'] = df_map['Mapping_Accuracy'].astype(float)
```

### Visualize Structural Variant Counts by Population

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=df_sv, x='Population', y='SV_Count', palette='Blues_d')
plt.title('Structural Variant Counts by Population')
plt.xlabel('Population')
plt.ylabel('Number of Structural Variants')
plt.tight_layout()
plt.show()
```

### Visualize Mapping Accuracy (GRCh38 vs Pangenome)

```python
plt.figure(figsize=(6, 4))
sns.barplot(data=df_map, x='Reference', y='Mapping_Accuracy', palette=['gray', '#1d434e'])
plt.title('Read Mapping Accuracy: GRCh38 vs Pangenome')
plt.ylabel('Alignment Accuracy (%)')
plt.ylim(0, 100)
plt.tight_layout()
plt.show()
```

### PCA Visualization of Population Genetic Variation

```python
plt.figure(figsize=(7, 5))
sns.scatterplot(data=df_pca, x='PC1', y='PC2', hue='Population', palette='Set2', s=100)
plt.title('PCA of Genetic Variation Across Populations')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.axhline(0, color='gray', linestyle='--')
plt.axvline(0, color='gray', linestyle='--')
plt.legend(title='Population')
plt.tight_layout()
plt.show()
```

 ### ðŸ§  Reflections

- The African population shows the highest number of structural variants, confirming greater genetic diversity.
- The pangenome provides better mapping accuracy than GRCh38, especially for diverse genomes.
- PCA shows how different populations cluster genetically, which supports the need for population-specific references.
